```python
def pairplot(df, target):
    ncol, nrow = 7, df.shape[1] // 7 + (df.shape[1] % 7 > 0)
    plt.figure(figsize=(ncol * 4, nrow * 4))

    for i, feature in enumerate(df.columns):
        plt.subplot(nrow, ncol, i + 1)
        plt.scatter(df[feature], target, s=10, marker='o', alpha=.6)
        plt.xlabel(feature)
        if i % ncol == 0:
            plt.ylabel('target')
```

Для одной комбинаций метода масштабирования и линейной регрессии постройте зависимость метрики $R^2$ на кросс-валидации (`GridSearchCV.cv_results_['mean_test_score']`) от значения параметра регуляризации для различных степеней полиномов в `PolynomialFeatures`.

```python
r2_score_plot= []
standard_pipeline=Pipeline([
    ('poly_feat', PolynomialFeatures(5)),
    ('std_scaler', MinMaxScaler()),
    ('cv', GridSearchCV(Lasso(max_iter=6000), parameters, n_jobs=-1, cv=5, error_score=r2_score, return_train_score=True)),    
])
for degree in range(1, 5):
    standard_pipeline.set_params(poly_feat__degree=degree)
    standard_pipeline.fit(x_train, np.log(y_train))
    r2_score_plot.append(standard_pipeline.named_steps['cv'].cv_results_['mean_test_score'])
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))

for i in range(len(r2_score_plot)):
    ax1.semilogx(parameters['alpha'][:30], r2_score_plot[i][:30], label=u"%d степень полинома" % (i+1))
    ax2.semilogx(parameters['alpha'][:10], r2_score_plot[i][:10], label=u"%d степень полинома" % (i+1))
    
ax1.set_title(u"График зависимости R^2 от альфа")
ax1.set_xlabel(u"Alphas")
ax1.set_ylabel(u"R^2 метрика")
ax1.grid()
ax1.legend()
ax2.set_title(u"Увеличенный график зависимости R^2 от альфа")
ax2.set_xlabel(u"Alphas")
ax2.set_ylabel(u"R^2 метрика")
ax2.grid()
ax2.legend()
f.show()
# Так как при слишком больших альфа не видно изменений метрики, я сделал 2 графики, первый показывает 
# общую зависимость метрики от альфа, а второй при маленьких альфа, где и происходят измения метрики 
# Использовал semilogx, что бы лучше видно на графике, ведь значения альфа слишком большие
```
